```python
import pandas as pd
```

```python
df=pd.read_csv('athlete_events.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
df
```

```python
df2=df.dropna()
df2
```

```python
import pandas as pd

# Specify the year before which you want to delete data
target_year = 1990

# Create a boolean mask to select rows with 'Year' greater than or equal to target_year
mask = df['Year'] >= target_year

# Use the mask to filter the DataFrame and keep rows with 'Year' greater than or equal to target_year
df3 = df2[mask]
```

```python
df4=df3[['Name','Sex','Season','Year','Sport', 'Medal','Team']].sort_values(by='Sport', ascending=True)
df4
```

```python
df4["ColorMedal"] = df4["Medal"]
```

```python
df4["ColorMedal"]= df4["ColorMedal"].replace('Bronze', "peru").apply(str.lower)
```

```python
NoMale=df4[df4['Sex']!='M']
NoMale
```

```python
WinterFemale=NoMale[(NoMale['Season'] == 'Winter')&(NoMale['Sex'] == 'F')]
WinterFemale
```

```python
WinterMale=df4[(df4['Season'] == 'Winter')&(df4['Sex'] == 'M')]
WinterMale
```

```python
SummerFemale=NoMale[(NoMale['Season'] == 'Summer')&(NoMale['Sex'] == 'F')]
SummerFemale
```

```python
SummerMale=df4[(df4['Season'] == 'Summer')&(df4['Sex'] == 'M')]
SummerMale
```

```python
SF = SummerFemale['Sport'].value_counts().nlargest(n=5)
SF
```

```python
top_sports = SummerFemale['Sport'].value_counts().nlargest(n=5).index
SF = SummerFemale[SummerFemale['Sport'].isin(top_sports)].copy()
SF
```

```python
SF.plot.scatter(x = 'Sport', y = 'Team',figsize=(10,20),c='ColorMedal');
plt.savefig('summerfemale.png',bbox_inches='tight')
```
